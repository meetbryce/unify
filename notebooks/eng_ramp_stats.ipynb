# Calculate engineer ramp time
By looking at the time delta between starting and commmiting their first PR

### get employee start date from our spreadsheet

```text
select login, members.name, Start, Level from github.users members
  join gsheets.employees emps on members.name = emps.Name 
  where Level < 7
  
```

```text
with members as (
  select login, members.name, Start, Level from github.users members
  join gsheets.employees emps on members.name = emps.Name 
  where Level < 7
)
select name,Start,merged_at, title, Level, (greatest(merged_at,Start) - Start) as ramp from members join (
  select title, html_url, merged_at, user_login 
    from github.merged_prs 
    where id in (select min(id) from github.merged_prs group by user_login)
) as early_pull
  on members.login = early_pull.user_login
  
```

```text
select name,Start, cast(Level as INT) as Level from gsheets.employees order by Start
```

```text
create chart as bar_chart where x = Start and y = level
```
