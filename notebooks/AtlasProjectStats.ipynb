## Atlas project analysis

export Atlas project list as 'projects.csv' and import it: import projects.csv

now make a view to filter down to eng teams only:

```text
create view
  files.eng_projects
AS SELECT * from
  files.projects
where
  Team not like '%TVP%' and Team not like '%R&D' and 
  Team not like '%Science%' and Team not like '%Design%' and
  Team not in ('Product','Mach 8')
```

```text
select Team, cast(count(*) as int) as count from files.eng_projects group by Team
```

```text
create chart as pie_chart where theta = 'count:Q' and color = 'Team:N'
```

```text
select Team, Status, count(*)  
  from files.eng_projects 
  group by Team, Status
  order by Team
```
