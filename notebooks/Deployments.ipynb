```text
select count(*) as count, base_repo_name as repo, date_trunc('week',created_at) as wk 
  from github.pulls 
where 
  state = 'closed' and base_ref in ('main','master') and 
    base_repo_name in (select repo_name from gsheets.deploy_repos) and
    created_at >= '2021-01-01'
group by 
 base_repo_name, date_trunc('week',created_at) order by wk desc
having
 count > 1
```

```text
create chart as line_chart where x = wk and y = count and color = repo and title = 'All PRs to main/master'
```
