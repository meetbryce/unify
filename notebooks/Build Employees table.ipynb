## Build Employee table combining Github and GSheets lists

### Import Gsheets employee data

Should do this before running this notebook:
  - drop table if exists gsheets.employees
  - gsheets create table employees from 'Copy of 00_Hiring_Eng'

```text
select * from gsheets.employees where Name like 'Will%' limit 10
```

```text
select count(*) from (
    select login, gu.name 
      from github.users gu 
      right join gsheets.employees emp on 
      (
        gu.name = emp.Name
      )
      where login is not null
)
```

```text
$emps = select login, gu.name, emp.Level as lvl, cast(Start as date) as start, emp.Manager as manager
      from github.users gu 
      left join gsheets.employees emp on 
      (
        replace(ltrim(rtrim(lower(gu.name))), 'nicholas','nick') = replace(ltrim(rtrim(lower(emp.Name))), 'nicholas','nick')
        or login = emp.ghlogin
      )
   where manager is not null
   order by start
```

```text
drop view if exists github.coders
```

```text
create view github.coders as
  SELECT 
      login, lvl, name, start, manager
    from
      $emps
    where
      manager <> '' and cast(lvl as int) < 7
```

```text
select count(*) from github.coders
```
